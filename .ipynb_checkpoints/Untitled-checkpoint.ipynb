```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
import numpy as np
import numpy.linalg as LA
stopWords = stopwords.words('english')
```

```python
all_abs = df['abstract'][0:10].tolist()
```

```python
from sklearn.metrics.pairwise import linear_kernel
```

```python
def abs_cs(abs_list):#create similarity matrix
    res = []
    vectorizer = CountVectorizer(stop_words = stopWords)
    transformer = TfidfTransformer()
    trainVectorizerArray = vectorizer.fit_transform(abs_list).toarray()
    #trainTfidfVectorizer = TfidfVectorizer().fit_transform(all_abs).toarray()
    #print('Fit Vectorizer to train set', trainVectorizerArray)
    transformer.fit(trainVectorizerArray)
    tfidf_list = transformer.transform(trainVectorizerArray).toarray()
    for i in range(len(tfidf_list)):
        cosine_similarities = linear_kernel(trainVectorizerArray[i:i+1], trainVectorizerArray).flatten()
        res.append(cosine_similarities)
    return np.array(res)
```

```python
abs_cs(all_abs)
```

```python
import re
```

```python
def author2index(all_aut):#author：[paper_idx, ...]
    res_dict = {}
    for author in all_aut.items():
        #print(author)
        res = re.split(", ", author[1])[:-1]
        for a in res:
            if a not in res_dict:
                res_dict[a] = []
            res_dict[a].append(author[0])
    return res_dict
        #print(res)
        
```

```python
author_dict = author2index(all_aut)
```

```python
def abs_sim(abs1, abs2):#similarity of two abstract
    abs_list = [abs1, abs2]
    vectorizer = CountVectorizer(stop_words = stopWords)
    transformer = TfidfTransformer()
    trainVectorizerArray = vectorizer.fit_transform(abs_list).toarray()
    #trainTfidfVectorizer = TfidfVectorizer().fit_transform(all_abs).toarray()
    #print('Fit Vectorizer to train set', trainVectorizerArray)
    transformer.fit(trainVectorizerArray)
    tfidf_list = transformer.transform(trainVectorizerArray).toarray()
    cosine_similarities = linear_kernel(trainVectorizerArray[0:1], trainVectorizerArray[1:2]).flatten()
    return cosine_similarities
```

```python
abs_sim(all_abs[0], all_abs[0])
```

```python
def sim(abstract, author_pub_dict, df):
    res_dict = {}
    for author in author_pub_dict:
        all_abs = df.loc[author_pub_dict[author]]['abstract']
        for abst in all_abs:
            if author not in res_dict:
                res_dict[author] = 0
            res_dict[author] += abs_sim(abstract, abst)
    
    return res_dict
```

```python
sim(all_abs[0],author_dict, df)
```
